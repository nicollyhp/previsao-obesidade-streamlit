Determinar quais colunas contêm dados que precisam ser traduzidos.

```python
colunas_para_traduzir_valores = []

for coluna in dados_traduzidos.columns:
    # Verificar tipo de dado
    if dados_traduzidos[coluna].dtype == 'object':
        # Verificar nomes de colunas que indicam dados categóricos a serem traduzidos
        if coluna in ['Gênero', 'Histórico Familiar', 'Consumo de Alimento Altamente Calórico',
                      'Consumo de Alimento Entre Refeições', 'Fumante', 'Monitoramento de Consumo de Calorias',
                      'Consumo de Álcool', 'Meio de Transporte Utilizado', 'Nível de Obesidade']:
            colunas_para_traduzir_valores.append(coluna)

print("Colunas com valores a serem traduzidos:")
print(colunas_para_traduzir_valores)
```

Mapear valores para tradução
Para cada coluna identificada, listar os valores únicos e criar um dicionário de mapeamento de inglês para português.

```python
colunas_para_traduzir_valores = ['Gênero', 'Histórico Familiar', 'Consumo de Alimento Altamente Calórico',
                                 'Consumo de Alimento Entre Refeições', 'Fumante', 'Monitoramento de Consumo de Calorias',
                                 'Consumo de Álcool', 'Meio de Transporte Utilizado', 'Nível de Obesidade']

mapeamento_valores = {}

for coluna in colunas_para_traduzir_valores:
    valores_unicos = dados_traduzidos[coluna].unique()
    print(f"Valores únicos na coluna '{coluna}': {valores_unicos}")

    # De para das colunas
    if coluna == 'Gênero':
        mapeamento_valores[coluna] = {'Female': 'Feminino', 'Male': 'Masculino'}
    elif coluna == 'Histórico Familiar':
        mapeamento_valores[coluna] = {'yes': 'Sim', 'no': 'Não'}
    elif coluna == 'Consumo de Alimento Altamente Calórico':
        mapeamento_valores[coluna] = {'yes': 'Sim', 'no': 'Não'}
    elif coluna == 'Consumo de Alimento Entre Refeições':
        mapeamento_valores[coluna] = {'Sometimes': 'Às vezes', 'Frequently': 'Frequentemente', 'no': 'não', 'Always': 'Sempre'}
    elif coluna == 'Fumante':
        mapeamento_valores[coluna] = {'yes': 'Sim', 'no': 'Não'}
    elif coluna == 'Monitoramento de Consumo de Calorias':
        mapeamento_valores[coluna] = {'yes': 'Sim', 'no': 'Não'}
    elif coluna == 'Consumo de Álcool':
        mapeamento_valores[coluna] = {'no': 'Não', 'Sometimes': 'Às vezes', 'Frequently': 'Frequentemente', 'Always': 'Sempre'}
    elif coluna == 'Meio de Transporte Utilizado':
        mapeamento_valores[coluna] = {'Public_Transportation': 'Transporte Público', 'Walking': 'Caminhando', 'Automobile': 'Automóvel', 'Motorbike': 'Moto', 'Bike': 'Bicicleta'}
    elif coluna == 'Nível de Obesidade':
        mapeamento_valores[coluna] = {'Normal_Weight': 'Peso Normal', 'Overweight_Level_I': 'Sobrepeso Nível I',
                                      'Overweight_Level_II': 'Sobrepeso Nível II', 'Obesity_Type_I': 'Obesidade Tipo I',
                                      'Insufficient_Weight': 'Peso Insuficiente', 'Obesity_Type_II': 'Obesidade Tipo II',
                                      'Obesity_Type_III': 'Obesidade Tipo III'}

print("\nDicionário de mapeamento de valores criado:")
print(mapeamento_valores)
```

Identificar colunas relevantes e Normalizar ou categorizar valores
Selecionar as colunas que podem contribuir para uma métrica de estilo de vida saudável e transformar os valores das colunas selecionadas em uma escala comparável.

```python
# Colunas relevantes para a métrica de estilo de vida saudável
colunas_saudaveis = [
    'Consumo de Alimento Altamente Calórico',
    'Consumo de Vegetais em Refeições Principais',
    'Número de Refeições Principais',
    'Consumo de Alimento Entre Refeições',
    'Fumante',
    'Consumo de Água Diário',
    'Consumo de Álcool',
    'Nível de Obesidade', # Adicionada a coluna Nível de Obesidade
    'Frequência de Atividade Física',
]

# Mapear valores categóricos para numéricos (quanto menor o número, mais saudável)
# Certifique-se de que as chaves no mapeamento correspondem aos valores em português no DataFrame dados_traduzidos
mapeamento_numerico_saudavel = {
    'Consumo de Alimento Altamente Calórico': {'Não': 3, 'Sim': 1}, # Não=mais saudável, Sim=menos saudável
    'Consumo de Vegetais em Refeições Principais': {'Não': 4, 'Sim': 1}, # Não=mais saudável, Sim=menos saudável
    'Consumo de Alimento Entre Refeições': {'Não': 4, 'Às vezes': 3, 'Frequentemente': 2, 'Sempre': 1}, # Não=mais saudável, Sempre=menos saudável
    'Fumante': {'Não': 4, 'Sim': 1}, # Não=mais saudável, Sim=menos saudável
    'Consumo de Álcool': {'Não': 7, 'Às vezes': 4, 'Frequentemente': 3, 'Sempre': 0}, # Não=mais saudável, Sempre=menos saudável
    'Monitoramento de Consumo de Calorias': {'Não': 1, 'Sim': 4}, # Não=menos saudável, Sim=mais saudável (ajustado)
    'Histórico Familiar': {'Não': 3, 'Sim': 1}, # Não=mais saudável, Sim=menos saudável
    'Nível de Obesidade': { # Mapeamento para Nível de Obesidade (quanto menor o nível de obesidade, mais saudável)
        'Peso Insuficiente': 5,
        'Peso Normal': 10,
        'Sobrepeso Nível I': 5,
        'Sobrepeso Nível II': 3,
        'Obesidade Tipo I': 0,
        'Obesidade Tipo II': -5, # Obesidade Tipo II e III geralmente são consideradas menos saudáveis
        'Obesidade Tipo III': -10 # Mapeando para um valor menor para indicar menos saudável
    }
}


# Aplicar o mapeamento numérico para as colunas categóricas relevantes
for coluna, mapeamento in mapeamento_numerico_saudavel.items():
    if coluna in dados_traduzidos.columns: # Aplicar diretamente em dados_traduzidos
        # Aplicar o mapeamento e depois converter para numérico, lidando com erros
        dados_traduzidos[coluna + '_numerico'] = dados_traduzidos[coluna].replace(mapeamento)
        dados_traduzidos[coluna + '_numerico'] = pd.to_numeric(dados_traduzidos[coluna + '_numerico'], errors='coerce')


# Normalizar colunas numéricas (opcional) - não será necessário se usarmos os valores diretamente na soma.

display(dados_traduzidos.head())
```

Cálculo do nível de vida saudável e criação de nova coluna
Criar uma fórmula para combinar os valores das colunas selecionadas em uma única métrica de estilo de vida saudável e adicionar a métrica calculada como uma nova coluna no DataFrame.

```python
# Fórmula para combinar as métricas (exemplo: soma ponderada - ajuste os pesos conforme a importância)
# É importante notar que esta é apenas uma sugestão de fórmula.
# A definição exata de "estilo de vida saudável" e a ponderação de cada fator
# podem variar e devem ser baseadas em conhecimento da área ou análise mais aprofundada.

# Vamos somar as pontuações e valores ajustados das colunas relevantes.
# Uma pontuação mais alta nesta métrica representará um estilo de vida "mais saudável".

# Usando as colunas originais para as numéricas e as colunas '_numerico' para as categóricas mapeadas
dados_traduzidos['Métrica Estilo Saudável'] = (
    dados_traduzidos['Consumo de Alimento Altamente Calórico_numerico'] +
    dados_traduzidos['Consumo de Vegetais em Refeições Principais'] + # Coluna numérica original
    dados_traduzidos['Número de Refeições Principais'] + # Coluna numérica original
    dados_traduzidos['Consumo de Alimento Entre Refeições_numerico'] +
    dados_traduzidos['Fumante_numerico'] +
    dados_traduzidos['Consumo de Água Diário'] + # Coluna numérica original
    dados_traduzidos['Consumo de Álcool_numerico'] +
    dados_traduzidos['Nível de Obesidade_numerico'] + # Usando a coluna numérica do Nível de Obesidade
    dados_traduzidos['Frequência de Atividade Física'] # Coluna numérica original
    # Adicione outras colunas relevantes aqui se desejar, usando seus nomes originais ou '_numerico' se mapeadas
)

# Preencher valores NaN na coluna 'Métrica Estilo Saudável' (por exemplo, com a média)
dados_traduzidos['Métrica Estilo Saudável'] = dados_traduzidos['Métrica Estilo Saudável'].fillna(dados_traduzidos['Métrica Estilo Saudável'].mean())


# Classificar em 5 níveis
# Podemos usar quantis para dividir a métrica em 5 grupos de tamanho aproximadamente igual
dados_traduzidos['Nível de Estilo de vida saudável'] = pd.qcut(dados_traduzidos['Métrica Estilo Saudável'], 5, labels=['Muito baixo', 'Baixo', 'Médio', 'Alto', 'Muito Alto'])


# Exibir as primeiras linhas com a nova coluna
display(dados_traduzidos.head())
```

Machine Learning - Random Forest

```python

# Importando bibliotecas
from sklearn.model_selection import train_test_split

# Copiando a base original
df_rf = df_ml.copy()

# Convertendo colunas object para category (opcional)
object_columns = df_rf.select_dtypes(include='object').columns
df_rf[object_columns] = df_rf[object_columns].astype('category')

# Separando X e y
X_rf = df_rf.drop(['Nível de Obesidade'], axis=1)
y_rf = df_rf['Nível de Obesidade']

# Convertendo variáveis categóricas para dummies SEM drop_first
X_rf = pd.get_dummies(X_rf, drop_first=False)

# Dividindo em treino e teste
X_treino_rf, X_teste_rf, y_treino_rf, y_teste_rf = train_test_split(
    X_rf, y_rf, test_size=0.2, random_state=45
)

# Verificando as dimensões
print(X_treino_rf.shape, X_teste_rf.shape, y_treino_rf.shape, y_teste_rf.shape)
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Criar cópias dos dados de treino e teste para evitar modificar os originais
X_treino_rf_encoded = X_treino_rf.copy()
X_teste_rf_encoded = X_teste_rf.copy()

# --- Remover IMC/BMI se existirem ---
colunas_para_excluir = [c for c in ['IMC', 'BMI'] if c in X_treino_rf_encoded.columns]
if colunas_para_excluir:
    X_treino_rf_encoded = X_treino_rf_encoded.drop(columns=colunas_para_excluir)
    X_teste_rf_encoded = X_teste_rf_encoded.drop(columns=colunas_para_excluir)

# Identificar colunas categóricas
categorical_features = X_treino_rf_encoded.select_dtypes(include=['object', 'category']).columns

# Aplicar Label Encoding às colunas categóricas nos conjuntos de treino e teste
for col in categorical_features:
    le = LabelEncoder()
    # Fit no conjunto de treino e transformar ambos
    X_treino_rf_encoded[col] = le.fit_transform(X_treino_rf_encoded[col])
    # Atenção: se houver categorias novas no teste, isso pode quebrar.
    # Aqui, garantimos que valores desconhecidos sejam mapeados para -1:
    mapa = {cls: i for i, cls in enumerate(le.classes_)}
    X_teste_rf_encoded[col] = X_teste_rf_encoded[col].map(mapa).fillna(-1).astype(int)

# Utilizando floresta para treinamento do modelo
rf = RandomForestClassifier(
    criterion='entropy',
    n_estimators=80,
    max_depth=7,
    class_weight='balanced',
    random_state=7
)
rf.fit(X_treino_rf_encoded, y_treino_rf)
```

```python

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, mean_squared_error
from sklearn.preprocessing import LabelEncoder

#Predições
y_predito_random_forest = rf.predict(X_teste_rf_encoded)
y_predito_random_forest_treino = rf.predict(X_treino_rf_encoded)

#Acurácia (treino e teste)
accuracy_treino = accuracy_score(y_treino_rf, y_predito_random_forest_treino)
accuracy_teste = accuracy_score(y_teste_rf, y_predito_random_forest)

print(f"Acurácia no conjunto de treino (Random Forest): {accuracy_treino:.2%}")
print(f"Acurácia no conjunto de teste (Random Forest): {accuracy_teste:.2%}")

# Ajusta o encoder nas classes do TREINO (as que o modelo conhece)
le_y = LabelEncoder()
le_y.fit(y_treino_rf)

# Converte y_teste e y_predito em valores numéricos de classe
y_true_num = le_y.transform(y_teste_rf)

# Mapeamento robusto para predições (evita erro se aparecer classe inesperada)
mapa_classes = {cls: i for i, cls in enumerate(le_y.classes_)}
y_pred_num = pd.Series(y_predito_random_forest).map(mapa_classes).fillna(-1).astype(int).to_numpy()

# Calcula MSE
mse_teste = mean_squared_error(y_true_num, y_pred_num)
print(f"\nMSE: {mse_teste:.4f}")

# Calcula RMSE
rmse_teste = mse_teste ** 0.5
print(f"RMSE: {rmse_teste:.4f}")
```

```python
#Verificando quais Feature possuem maior importância no aprendizado do Machine Learning
importances = rf.feature_importances_
feature_names = X_treino_rf_encoded.columns
forest_importances = pd.Series(importances, index=feature_names)

# Ordenar as features por importância
forest_importances = forest_importances.sort_values(ascending=True)

fig, ax = plt.subplots(figsize=(10, 6))
forest_importances.plot.barh(ax=ax)
ax.set_title("Importância das Features")
ax.set_ylabel("Média da Redução de Impureza")
plt.grid(False)
plt.show()
```

```python
# Separando uma cópia da variável df com as features originais da base de dados para utilizar no modelo
df_xgb = df_ml.copy()

# Transformando os dados do tipo Object para category
object_columns = df_xgb.select_dtypes(include='object').columns
df_xgb[object_columns] = df_xgb[object_columns].astype('category')

from sklearn.preprocessing import LabelEncoder

# Tratando o coluna alvo 'Nível de Obesidade' para que ela fique codificada
le = LabelEncoder()
df_xgb['Nível de Obesidade_encoded'] = le.fit_transform(df_xgb['Nível de Obesidade'])

# Convertendo a coluna alvo do tipo 'object' para XGBoost
X_xgb = df_xgb.drop(['Nível de Obesidade', 'Nível de Obesidade_encoded'], axis=1)

y_xgb = df_xgb['Nível de Obesidade_encoded']
```

```python
# Separando os grupos de Treino e Teste
from sklearn.model_selection import train_test_split

X_treino_xgb, X_teste_xgb, y_treino_xgb, y_teste_xgb = train_test_split(X_xgb, y_xgb, test_size=0.2, random_state=45)

#Treinando o modelo utilizando XGBoost
import xgboost as xgb

modelo_xgb = xgb.XGBRegressor(objective='reg:squarederror', enable_categorical=True, n_estimators=100)
# Change the model type to XGBClassifier for probability prediction
#modelo_xgb = xgb.XGBClassifier(objective='multi:softprob', enable_categorical=True, n_estimators=100)


modelo_xgb.fit(X_treino_xgb, y_treino_xgb)
preds = modelo_xgb.predict(X_teste_xgb)
```

```python
# Verificando as métricas MSE e RMSE
from sklearn.metrics import mean_squared_error, accuracy_score, classification_report
import numpy as np

mse = mean_squared_error(y_teste_xgb, preds)
print(f'MSE: {mse:.2f}')

rmse = mean_squared_error(y_teste_xgb, preds) ** 0.5
print(f'RMSE: {rmse:.2f}')

# Convertendo as previsões contínuas para classes discretas arredondando e convertendo para int
preds_rounded = np.round(preds).astype(int)
preds_treino_rounded = np.round(modelo_xgb.predict(X_treino_xgb)).astype(int) # convertendo a predição de treino

# Verificando a acurácia usando as previsões arredondadas
acc_treino = accuracy_score(y_treino_xgb, preds_treino_rounded)
acc_teste = accuracy_score(y_teste_xgb, preds_rounded)

print(f"Acurácia no conjunto de treino (XGBoost): {acc_treino:.2%}")
print(f"Acurácia no conjunto de teste (XGBoost): {acc_teste:.2%}")

# Exibir um relatório de classificação para obter métricas mais detalhadas
print("\nRelatório de Classificação:")
print(classification_report(y_teste_xgb, preds_rounded))
```
